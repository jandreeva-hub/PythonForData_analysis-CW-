### Биномиальное распределение

```python
from scipy.stats import binom
import seaborn as sns
import matplotlib.pyplot as plt
```

Пример одного испытания: подбрасывание монетки

```python
# Параметры
n = 10 # количество испытаний
p = 0.5 # вероятность успеха
```

```python
# Генерируем случайные величины с биномиальным распределением
data_binom = binom.rvs(n=n, p=p, size=10000)
```

```python
len(data_binom)
```

```python
# Генерируем случайные величины с биномиальным распределением
data_binom = binom.rvs(n=n, p=p, size=10000)

# Построение графика распределения
ax = sns.histplot(data_binom, kde=False, color='skyblue', bins=30)#, stat='probability')
ax.set(xlabel='Биномиальное распределение', ylabel='Частота')
plt.title(f'Биномиальное распределение (n={n}, p={p})')
plt.show()
```

### Распределение Пуассона

Рассмотрим сценарий, в котором колл-центр получает в среднем 3 звонка в минуту. 

Вы хотите найти вероятность получения ровно 5 звонков за минуту.

```python
from scipy.stats import poisson
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Параметр
mu = 3 # среднее число вхождений (лямбда)

# Генерируем случайные переменные с пуассоновским распределением
data_poisson = poisson.rvs(mu=mu, size=10000)
```

### Равномерное распределение

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Параметры
a = 0 # нижняя граница
b = 10 # верхняя граница

# Генерируем равномерно распределенные случайные величины
data_uniform = uniform.rvs(loc=a, scale=b-a, size=1000)
```

```python
# Построение графика распределения
plt.hist(data_uniform, bins=30, density=True, alpha=0.6, color='b')
plt.show()
```

### Асимметрия (skewness)

```python
import pandas as pd
import numpy as np

# Пример набора данных
data = np.random.normal(0, 1, 1000)

# Рассчёт ассиметрии с помощью pandas
skewness_pandas = pd.Series(data).skew()
print(f"Skewness (pandas): {skewness_pandas}")
```

### Эксцесс (kurtosis)

```python
# Вычисление эксцесса с помощью pandas
kurtosis_pandas = pd.Series(data).kurt()
print(f"Kurtosis (pandas): {kurtosis_pandas}")
```

### Дополнительный пример с распределением Пуассона (счет футбольного матча)

```python
import numpy as np
from scipy.stats import poisson
```

```python
home_goals = [2,3,3,1,3,2,0,2,1]
away_goals = [2,4,1,1,0,1,0,0,0]

home_avg = np.mean(home_goals)
away_avg = np.mean(away_goals)
```

```python
home_probs = [poisson.pmf(i, home_avg) for i in range(5)]
away_probs = [poisson.pmf(i, away_avg) for i in range(5)]
```

```python
match_probs = np.outer(home_probs, away_probs)
```

```python
for i in range(5):
    for j in range(5):
        print(f"Probability of {i}-{j} match score: {match_probs[i,j]:.3f}")
```
